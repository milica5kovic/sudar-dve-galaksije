```python
import numpy as np
import math
import matplotlib.pyplot as plt
import csv

path = "galaxy.txt"
m,x,y,z,vx,vy,vz = np.loadtxt(path, dtype='float', unpack = True) #pocetni uslovi za 400 cestica

m_d = []
m_b = []
m_h = []
x_d = []
x_b = []
x_h = []
y_d = []
y_b = []
y_h = []
z_d = []
z_b = []
z_h = []
vx_d = []
vx_b = []
vx_h = []
vy_d = []
vy_b = []
vy_h = []
vz_d = []
vz_b = []
vz_h = []

for i in range(100):
    m_d.append(m[i])
    x_d.append(x[i])
    y_d.append(y[i])
    z_d.append(z[i])
    vx_d.append(vx[i])
    vy_d.append(vy[i])
    vz_d.append(vz[i])

for i in range (100):
    m_b.append(m[i+100])
    x_b.append(x[i+100])
    y_b.append(y[i+100])
    z_b.append(z[i+100])
    vx_b.append(vx[i+100])
    vy_b.append(vy[i+100])
    vz_b.append(vz[i+100])

for i in range (200):
    m_h.append(m[i+200])
    x_h.append(x[i+200])
    y_h.append(y[i+200])
    z_h.append(z[i+200])
    vx_h.append(vx[i+200])
    vy_h.append(vy[i+200])
    vz_h.append(vz[i+200])

class telo():
    def __init__(self, m, x, y, z, vx, vy, vz, ax, ay, az, marker):
        self.m = m
        self.x = x
        self.y = y
        self.z = z
        self.vx = vx
        self.vy = vy
        self.vz = vz
        self.ax = ax
        self.ay = ay
        self.az = az
        self.marker = marker


def jedinice_masa(m):
    return m*2.325e9

def jedinice_brzina(v):
    return v*3.24e-15
  
def rastojanje(t1,t2):
    return math.sqrt((t1.x-t2.x)**2+(t1.y-t2.y)**2 + (t1.z-t2.z)**2)

def main():
    
    N = 800
    dt = 1e7*365*86400
    G = 4.53e-39
    tmax = 1e9*365*86400
    t = 0
    snapshot = 0
    E = []
    T = []
    l = 200
    vp = 100

    tela = []
    
    for i in range(200):
        tela.append(telo(jedinice_masa(m_h[i]), x_h[i]+l, y_h[i], z_h[i], jedinice_brzina(vx_h[i] - vp/100), jedinice_brzina(vy_h[i]), jedinice_brzina(vz_h[i]), 0, 0, 0, "halo"))
        tela.append(telo(jedinice_masa(m_h[i]),x_h[i] , y_h[i], z_h[i], jedinice_brzina(vx_h[i]), jedinice_brzina(vy_h[i]), jedinice_brzina(vz_h[i]), 0, 0, 0, "halo1"))
    for i in range(100):
        tela.append(telo(jedinice_masa(m_d[i]), x_d[i]+l, y_d[i], z_d[i], jedinice_brzina(vx_d[i] - vp/100), jedinice_brzina(vy_d[i]), jedinice_brzina(vz_d[i]), 0, 0, 0, "disk")) 
        tela.append(telo(jedinice_masa(m_d[i]), x_d[i], y_d[i], z_d[i], jedinice_brzina(vx_d[i]), jedinice_brzina(vy_d[i]),jedinice_brzina(vz_d[i]), 0, 0, 0, "disk1"))
    for i in range(100):
        tela.append(telo(jedinice_masa(m_d[i]), x_b[i]+l, y_b[i], z_b[i], jedinice_brzina(vx_b[i] - vp/100), jedinice_brzina(vy_b[i]), jedinice_brzina(vz_b[i]), 0, 0, 0, "bulge"))
        tela.append(telo(jedinice_masa(m_b[i]), x_b[i] ,y_b[i], z_b[i], jedinice_brzina(vx_b[i]), jedinice_brzina(vy_b[i]), jedinice_brzina(vz_b[i]), 0, 0, 0, "bulge1"))
    
    #plt.scatter(x_h, y_h)
    #plt.scatter(x_d, y_d)
    #plt.scatter(x_b, y_b)
    #plt.show()
    
    x1_h = []
    y1_h = []
    z1_h = []

    x1_d = []
    y1_d = []
    z1_d = []
    
    x1_b = []
    y1_b = []
    z1_b = []
    
    x2_h = []
    y2_h = []
    z2_h = []

    x2_d = []
    y2_d = []
    z2_d = []
    
    x2_b = []
    y2_b = []
    z2_b = []
    
    
    
    while(t<tmax):

        if snapshot%100==0:
            ime_snapshota = 'ss' + str(snapshot)
            
            data = []
            
            for i in range(N):
                data.append([tela[i].m, tela[i].x, tela[i].y, tela[i].z, tela[i].vx, tela[i].vy, tela[i].vz])

            #with open(ime_snapshota, 'w', encoding='UTF8', newline='') as f:
                #writer = csv.writer(f)
                #writer.writerows(data)
                
        snapshot += 1
        
        e_pot = 0
        e_kin = 0
        
        for i in range(N):
            
            e_kin += tela[i].m * (tela[i].vx**2 + tela[i].vy**2 + tela[i].vz**2) / 2
            
            for j in range(N):
                #print(i,j)
                if (i!=j):
                    
                    d = rastojanje(tela[i],tela[j])
                    
                    e_pot += -G * tela[i].m * tela[j].m / d
                    
                    tela[i].ax += -G * tela[j].m * (tela[i].x-tela[j].x) / ((d)**3)
                    tela[i].ay += -G * tela[j].m * (tela[i].y-tela[j].y) / ((d)**3)
                    tela[i].az += -G * tela[j].m * (tela[i].z-tela[j].z) / ((d)**3)
            
            

        for i in range(N):
            
            tela[i].vx += dt * tela[i].ax
            tela[i].vy += dt * tela[i].ay
            tela[i].vz += dt * tela[i].az

            tela[i].x += dt * tela[i].vx
            tela[i].y += dt * tela[i].vy
            tela[i].z += dt * tela[i].vz
            
        E.append(e_kin+e_pot)
        T.append(t)
            
        if(t%(1e8*365*86400)==0):
            
            print(t/tmax*100)
            
            x1_h = []
            y1_h = []
            z1_h = []
            x1_d = []
            y1_d = []
            z1_d = []
            x1_b = []
            y1_b = []
            z1_b = []
            x2_h = []
            y2_h = []
            z2_h = []
            x2_d = []
            y2_d = []
            z2_d = []
            x2_b = []
            y2_b = []
            z2_b = []
            
            for i in range(N):
                
                if (tela[i].marker == 'halo'):
                    x1_h.append(tela[i].x)
                    y1_h.append(tela[i].y)
                    z1_h.append(tela[i].z)
                if (tela[i].marker == 'disk'):
                    x1_d.append(tela[i].x)
                    y1_d.append(tela[i].y)
                    z1_d.append(tela[i].z)
                if (tela[i].marker == 'bulge'):
                    x1_b.append(tela[i].x)
                    y1_b.append(tela[i].y)
                    z1_b.append(tela[i].z)
                if (tela[i].marker == 'halo1'):
                    x2_h.append(tela[i].x)
                    y2_h.append(tela[i].y)
                    z2_h.append(tela[i].z)
                if (tela[i].marker == 'disk1'):
                    x2_d.append(tela[i].x)
                    y2_d.append(tela[i].y)
                    z2_d.append(tela[i].z)
                if (tela[i].marker == 'bulge1'):
                    x2_b.append(tela[i].x)
                    y2_b.append(tela[i].y)
                    z2_b.append(tela[i].z)
            
            plt.scatter(x1_h, y1_h, s=2)
            plt.scatter(x1_d, y1_d, s=2)
            plt.scatter(x1_b, y1_b, s=2)
            plt.scatter(x2_h, y2_h, s=2)
            plt.scatter(x2_d, y2_d, s=2)
            plt.scatter(x2_b, y2_b, s=2)
            plt.xlabel('X[kpc]')
            plt.ylabel('T[s]')
            plt.legend(['halo1','disk1','bulge1','halo2','disk2','bulge2'],loc = 'upper left')
            plt.xlim(-200, 350)
            plt.ylim(-165, 200)
            print("grafik u ",t)
            ime_grafika = 'grafik u trenutku ' + str(t) + '.png'
            plt.savefig(ime_grafika)
            #plt.title()
            
            
            plt.show()

            
            #print(len(x1_b))
        
        #for i in range(N):
            #print(tela[i].ax,tela[i].ay, tela[i].az, "\n")
        
        for i in range(N):
            tela[i].ax = 0
            tela[i].ay = 0
            tela[i].az = 0
            
        
        #print(100.0*t/tmax)
        t += dt

    
    plt.plot(T,E)
    plt.xlabel("t[s]")
    plt.ylabel("E[MâŠ™*kpc**2/s**2]")
    plt.title("Zavisnost ukupne energije sistema od vremena")
    plt.savefig("Zavisnost energije sistema od vremena")
    plt.show()

    
main()
```
